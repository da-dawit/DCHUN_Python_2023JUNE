```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("annual_gold_rate.csv")
df.head(5)
```

```python
sortedDf = df.sort_values(by='Date')
```

```python
x = sortedDf['Date']
y = sortedDf['USD']

plt.bar(x, y)
plt.show()
```

```python
plt.plot(x,y, label="USD", linestyle="solid")

a1 = sortedDf["EUR"]
plt.plot(x, a1, label="EUR", linestyle="dotted")

a2 = sortedDf["GBP"]
plt.plot(x,a2, label="GBP", linestyle="dashdot")

plt.legend()
plt.title("Date vs Gold Currency")
plt.xlabel("DATE by year")
plt.ylabel("Currency Value")
plt.tight_layout()
plt.show()
```
